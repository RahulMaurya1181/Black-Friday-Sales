## Cleaning and preparing the data for model training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# importing the dataset
df_train=pd.read_csv("blackfriday_train.csv")
df_test=pd.read_csv("blackfriday_test.csv")
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_train.describe()
```

```python
df_test.shape
```

```python
# Merge of two datasets
df=pd.concat([df_train,df_test], axis=0)
```

```python
df
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df
```

```python
## Handling the Categorical Feature Gender
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
## Handling the Categorical Feature Age
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True,dtype=float)
```

```python
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
```

```python
df.head()
```

```python
## Droping out the City Category
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
## Missing Values
df.isnull().sum()
```

```python
## Product_category_2 Replacing Missing Values
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
df['Product_Category_2'].mode()[0]
```

```python
## Replace the missing value with mode
df['Product_Category']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category'].isnull().sum()
```

```python
## Product_category_3 Replacing Missing Values
df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'].value_counts()
```

```python
df['Product_Category_3'].mode()[0]
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df['Stay_In_Current_City_Years'].unique()
```
